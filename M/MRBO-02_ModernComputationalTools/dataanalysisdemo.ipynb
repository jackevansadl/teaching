```python
# import required libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the a time series
t = np.linspace(0, 5000, num=10000)
```

```python
# generate some data to analyse
sig =  np.sin(30 * 2 * np.pi * t)
# cumulative sum of n random numbers
```

```python
# plot this
plt.plot(t, sig, color="C0")
```
