```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
H1 = ["H", np.array([1.1610300000,      0.0660680000,      1.0238850000  ])]
C1 = ["C", np.array([0.6578960000,     -0.0044590000,      0.0639790000])]
H2 = ["H", np.array([1.3351900000,     -0.0830130000,     -0.7814260000 ])]
C2 = ["C", np.array([ -0.6578350000,      0.0044580000,     -0.0638470000])]
H3 = ["H", np.array([-1.3354850000,      0.0830070000,     0.7812720000])]
H4 = ["H", np.array([-1.1607960000,     -0.0660610000,     -1.0238630000 ])]
```

```python
ethene = [H1,C1,H2,C2,H3,H4]
```

```python
for atom in ethene:
    print(atom)
```

```python
# plot a 2D representation
for atom in ethene:
    if atom[0] == 'H':
        plt.plot(atom[1][0], atom[1][2],'o', markersize=20, color="C6")
    elif atom[0] == 'C':
        plt.plot(atom[1][0], atom[1][2],'o', markerssize=20, color='k')
plt.xlabel("x / Å")
plt.ylabel("y / Å")
```
