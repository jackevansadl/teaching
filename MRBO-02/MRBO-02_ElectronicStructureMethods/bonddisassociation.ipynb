```python
import numpy
from pyscf import scf
from pyscf import gto



ehf = []

def run(b, dm):
    mol = gto.Mole()
    mol.atom = [
        ["H", (0., 0., 0.)],
        ["H", (0., 0., b)],]

    mol.basis = {'H': 'cc-pvdz',
                 'H': 'cc-pvdz',}
    mol.build()
    mf = scf.RHF(mol)
    ehf.append(mf.scf(dm))

    return mf.make_rdm1()

dm = None
for b in numpy.arange(0.4, 2.01, 0.05):
    dm = run(b, dm)

```

```python
x = numpy.arange(0.4, 2.01, .05)

import matplotlib.pyplot as plt
plt.plot(x, ehf)
```
