```python
# reproduced from http://nznano.blogspot.com/2018/03/simple-quantum-chemistry-hartree-fock.html

# Need to import some libraries to have the maths functions and plotting
import numpy as np
import scipy.special as sp
import matplotlib.pyplot as plt

x = np.linspace(-5,5,num=1000)
r = abs(x)


x = np.linspace(-5,5,num=1000)
r = abs(x)

zeta = 1.0

chi_STO = (zeta**3/np.pi)**(0.5)*np.exp(-zeta*r)
plt.xlabel("r")
plt.ylabel("$\chi$")

plt.plot(x,chi_STO)
```

```python
Coeff = np.array([[1.00000,0.0000000,0.000000],
                  [0.678914,0.430129,0.000000],
                  [0.444635,0.535328,0.154329]])

# Expon is the alpha variable in the equation above
Expon = np.array([[0.270950,0.000000,0.000000],
                  [0.151623,0.851819,0.000000],
                  [0.109818,0.405771,2.227660]]) 

chi_CGF_STO1G = Coeff[0,0]*(2*Expon[0,0]/np.pi)**(0.75)*np.exp(-Expon[0,0]*r**2)
chi_CGF_STO2G = Coeff[1,0]*(2*Expon[1,0]/np.pi)**(0.75)*np.exp(-Expon[1,0]*r**2) \
                + Coeff[1,1]*(2*Expon[1,1]/np.pi)**(0.75)*np.exp(-Expon[1,1]*r**2) \
                + Coeff[1,2]*(2*Expon[1,2]/np.pi)**(0.75)*np.exp(-Expon[1,2]*r**2)
chi_CGF_STO3G = Coeff[2,0]*(2*Expon[2,0]/np.pi)**(0.75)*np.exp(-Expon[2,0]*r**2) \
                + Coeff[2,1]*(2*Expon[2,1]/np.pi)**(0.75)*np.exp(-Expon[2,1]*r**2) \
                + Coeff[2,2]*(2*Expon[2,2]/np.pi)**(0.75)*np.exp(-Expon[2,2]*r**2)
    
# Plot the three functions
plt.figure(figsize=(3,3))
plt.plot(x,chi_STO,label="STO")
plt.plot(x,chi_CGF_STO1G,label="STO-1G")
plt.xlabel("r")
plt.ylabel("$\chi$")
plt.legend()
plt.figure(figsize=(3,3))
plt.plot(x,chi_STO,label="STO")
plt.xlabel("r")
plt.ylabel("$\chi$")
plt.plot(x,chi_CGF_STO2G,label="STO-2G")
plt.legend()
plt.figure(figsize=(3,3))
plt.plot(x,chi_STO,label="STO")
plt.xlabel("r")
plt.ylabel("$\chi$")
plt.plot(x,chi_CGF_STO3G,label="STO-3G")
plt.legend()
```
