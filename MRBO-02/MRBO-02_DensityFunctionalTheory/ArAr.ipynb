```python
import numpy
from pyscf import scf
from pyscf import gto
from pyscf import dft



ehf = []

def run(b, dm):
    mol = gto.Mole()
    mol.atom = [
        ["Ar", (0., 0., 0.)],
        ["Ar", (0., 0., b)],]

    mol.basis = {'Ar': 'cc-pvdz',
                 'Ar': 'cc-pvdz',}
    mol.build()
    mf = scf.RHF(mol)
    ehf.append(mf.scf(dm))

    return mf.make_rdm1()

dm = None
for b in numpy.arange(3.0, 8.1, 0.1):
    dm = run(b, dm)

```

```python
x = numpy.arange(3.0, 8.1, 0.1)

import matplotlib.pyplot as plt
plt.plot(x, ehf)
```
